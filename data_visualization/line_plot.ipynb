# Import Libraries
- Seaborn automatically installs thes libraries
* numpy
* scipy
* pandas
* matplotlib

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
phool = sns.load_dataset("iris")
phool
```

```python
#draw line plot
sns.lineplot(data = phool, x="sepal_length", y="sepal_width")
plt.title("Pholon Ka Line plot")
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data set
kashti = sns.load_dataset("titanic")
kashti
```

```python
#draw line plot
sns.barplot(data = kashti, x="sex", y="alone", hue="who", order=["female","male"])
plt.title("Titanic Ka Bar plot")
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median

#load data set
kashti = sns.load_dataset("titanic")
kashti

#draw line plot
sns.barplot(data = kashti, x="sex", y="alone", hue="who", order=["female","male"], color="coral", ci=None,estimator=median)
plt.title("Titanic Ka Bar plot")
plt.show()
```
