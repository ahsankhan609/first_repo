```python
# data in specific row position
df.iloc[0:5, 0:1] # 1st is row 2nd is column
```

```python
# data in specific column position
df.iloc[:, 0:1] # 1st is row 2nd is column, shows all data in the specific column.
```

```python
# show values using boolean in the specific column
df[df["A"] > 1.0] #shows values within the specific column
#df[df > 2] # shows values from all over the dataset
```
